```python
x=df.displacement
def minmax(x):
    return (x-min(x))/(max(x)-min(x))
print(minmax(x))
sns.histplot(minmax(x))
```

```python
def f_z_score(x):
    mean = np.mean(x)
    std = np.std(x)
    z_scores = [(y-mean)/std for y in x]
    return z_scores
print(f_z_score(x))
sns.histplot(f_z_score(x))
```
