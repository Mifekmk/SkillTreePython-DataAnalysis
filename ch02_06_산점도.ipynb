### (1) 데이터에 따른 크기 및 색깔 지정


````python
n = 50
np.random.seed(42) # 시드지정
x = np.random.rand(n) # 균등분포
y = np.random.rand(n) # 균등분포
x.shape, y.shape
````

````python
plt.scatter(x,y, #x/y 길이와 똑같아야함
            c = x,
            cmap='rainbow',
            )
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
````

````python
plt.scatter(x,y, #x/y 길이와 똑같아야함
            c = y,
            cmap='rainbow',
            )
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
````

````python
plt.scatter(x,y, #x/y 길이와 똑같아야함
            c = x*y,
            cmap='rainbow',
            )
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
````

````python
# size
plt.scatter(x,y, #x/y 길이와 똑같아야함
            c = x,
            s = x*500,
            cmap='rainbow',
            alpha = 0.8
            )
plt.colorbar() # 컬러바 표기
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
````

### (2) 산점도를 이용한 아이리스 데이터 시각화



```
           꽃받침 길이   꽃받침 너비   꽃의 길이  꽃의 너비
 샘플1      
 샘플2 
 ...
 샘플150
```


```
from sklearn.datasets import load_iris
iris = load_iris()
```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()

x_names = iris.feature_names
y_names = iris.target_names

x_data = iris.data
y_target = iris.target
x_names
# 시각화
plt.scatter(x_data[:,2],x_data[:,3],
            c=y_target,
            s=(x_data[:,0]*x_data[:,1])**2,
            alpha=0.4,
            cmap=plt.cm.get_cmap('rainbow',3)) #컬러바 3개로 표현
plt.xlabel(x_names[2])
plt.ylabel(x_names[3])
plt.title('Species of petal size')
plt.colorbar(ticks=[0,1,2], label='Species') #컬러바 3개로 표현
plt.clim(-0.5, 2.5)
plt.show()
# sepal width, length -> x_data[:,0],[:,1]
````
