### **| 예측값에 대한 추가 시각화**

* `countplot`으로 그리기


```python
f, axes = plt.subplots(1,len(noms), figsize=(30,6))
axes = axes.flatten()
for col,ax in zip(noms, axes):
    sns.countplot(data = df, x=col, ax=ax, alpha=0.6,)
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

* `hue` 옵션에 범주형 속성 추가하기

```python
f, axes = plt.subplots(1,len(noms), figsize=(36,6))
axes = axes.flatten()
for col,ax in zip(noms, axes):
    sns.countplot(data = df, x=col, ax=ax, alpha=0.6,hue='survived')
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
sns.pairplot(df.sample(100), hue='survived')
```

* `displot` 이용해 연속 자료형 별 예측값의 분포 `kde` 선으로 시각화 하기 

```python
for col in nums:
    sns.displot(data=df, x=col, kind='kde',hue='survived')
    plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `sex` 추가하기

```python
for col in nums:
    sns.displot(data=df, x=col, kind='kde',hue='survived',col='sex')
    plt.show()
```

* 위에서 그린 플롯에 `hue` 옵션으로 `class` 추가하기

```python
for col in nums:
    sns.displot(data=df, x=col, kind='kde',hue='class',col='sex')
    plt.show()
```

* `age`과 `fare`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x='age', y ='fare',hue='survived')
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `class` 추가하기

```python
plt.figure(figsize=(15,8))
sns.lmplot(data=df, x='age', y ='fare',hue='survived',col='class')
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
fig, axes = plt.subplots(1, len(noms), figsize=(36,5))
for col, ax in zip(noms, axes):
    _=sns.histplot(data=df, x=col, y='survived' ,ax=ax)
```
