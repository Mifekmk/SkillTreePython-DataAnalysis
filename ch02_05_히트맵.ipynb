### (1) pivot 테이블 이용한 시계열 데이터 시각화


```
np.random.seed(42)
t = ['오전','오후','저녁']*5
day=[f'2020/01/0{i+1}' for i in range(5) for j in range(3)]
vals = np.random.randint(100,500, len(t))

```

````python
np.random.seed(42)
t = ['오전','오후','저녁']*5
day=[f'2020/01/0{i+1}' for i in range(5) for j in range(3)]
vals = np.random.randint(100,500, len(t))

df = pd.DataFrame([day, t]).T # 리스트에 묶이면 vals 값이 object 형식으로 바뀝니다.
df.columns = ['day', 'time']
df['count'] = vals
df.head()
df.info()
df
````

````python
df.pivot('day', 'time', 'count').style.background_gradient()
````

### (2) Iris 데이터 변수별 상관관계 분석


```
           꽃받침 길이   꽃받침 너비   꽃의 길이  꽃의 너비
 샘플1      
 샘플2 
 ...
 샘플150
```


```
from sklearn.datasets import load_iris
iris = load_iris()
```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()

x_names = iris.feature_names
y_names = iris.target_names

x_data = iris.data
y_target = iris.target
# 데이터 프레임 만들기
df = pd.DataFrame(x_data)
df.columns= x_names
df['label']=y_target
c = df.corr()
````

````python
c
````

````python
#sns로 만들기
from sklearn.datasets import load_iris
iris = load_iris()

x_names = iris.feature_names
y_names = iris.target_names

x_data = iris.data
y_target = iris.target
# 데이터 프레임 만들기
df = pd.DataFrame(x_data)
df.columns= [x_names]
df['label']=y_target
df.corr()

sns.heatmap(df.corr(), annot=True)
````

````python
c.style.background_gradient()
````
