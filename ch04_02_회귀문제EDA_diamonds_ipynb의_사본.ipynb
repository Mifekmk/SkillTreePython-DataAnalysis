### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
sns.pairplot(df.sample(100), hue='carat')
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
f, axes = plt.subplots(1,len(noms), figsize=(45,10))
axes = axes.flatten()
for col,ax in zip(noms, axes):
    sns.boxplot(data = df, x=col, y='price',ax=ax)
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(25,5))
axes = axes.flatten()
for col,ax in zip(nums, axes):
    sns.scatterplot(data = df, x=col, y='price',ax=ax)
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

* `hue` 옵션에 `color` 추가해서 그리기 

```python
f, axes = plt.subplots(1,len(nums), figsize=(25,5))
axes = axes.flatten()
for col,ax in zip(nums, axes):
    sns.scatterplot(data = df, x=col, y='price',ax=ax, hue='color')
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

* `regplot`으로 회귀 시각화

```python
f, axes = plt.subplots(1,len(nums), figsize=(30,6))
axes = axes.flatten()
for col,ax in zip(nums, axes):
    sns.regplot(data = df, x=col, y='price',ax=ax, marker='.') # hue 옵션은 제공하지 않습니다.
    # 범주값의 경우 countplot으로 만드는게 좋습니다.
plt.show()
```

* `residplot`으로 잔차 시각화

```python
f, axes = plt.subplots(1,len(nums), figsize=(30,6))
axes = axes.flatten()
for col,ax in zip(nums, axes):
    sns.residplot(data = df, x=col, y='price',ax=ax,)
plt.show()
```

* `lmplot`으로 `color`별 회귀 시각화

```python
for col,ax in zip(nums, axes):
    sns.lmplot(data = df, x=col, y='price' ,hue='color')
    plt.show()
```

```python
for col,ax in zip(nums, axes):
    sns.lmplot(data = df, x=col, y='price' ,hue='color',col='color')
    plt.show()
```
