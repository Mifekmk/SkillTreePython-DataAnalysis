```python
plt.figure(figsize=(35,13))
plt.rc('font',size=20)
sns.heatmap(data=data.corr(), annot=True,
            fmt='0.2f',linewidths=0.5, cmap='Blues')
# 서울, 경기의 집값을 기준으로 함께 오르는 지역을 확인해보았습니다. 수도권 지역 중 서북권과 서남권이 특히 동기간 때에 오른 것을 확인 할 수 있었습니다.
# 반면에 대구, 부산, 인천, 광주 등은 상대적으로 덜 비례한 것을 볼 수 있었습니다.
# 울산, 충북, 충남, 경북, 경남 등은 오히려 0에 더 가까운 수치들을 보여주고 있습니다. 일자리와 문화생활등의 사회적인 여파로 이촌향도 현상이 심각한 것을 알 수 있습니다.
plt.savefig('correlation.png')
```

```python
seoul = data.iloc[:,25]
daegu = data.iloc[:,5]

def f(seoul):
    return data.iloc[:,0]
def f(daegu):
    return data.iloc[:,0]
plt.figure(figsize=(60,15))
plt.grid(True)
plt.plot(f(seoul),seoul, label='SEOUL',color='red', marker='o')
plt.plot(f(daegu),daegu, label='DAEGU',color='blue', marker='x')
plt.xlabel('2006년 1월 1일 ~ 2021년 1월 1일 시간 경과', labelpad=15, fontsize=40)
plt.ylabel('아파트 실거래 가격 지수',labelpad=20, fontsize=40)
plt.title('서울과 대구의 단위 기간 별 아파트 실거래 가격 지수', fontsize=40)
plt.legend(fontsize=50, ncol=1, frameon=True, shadow=True, loc='upper left')
plt.savefig('SEOUL and DAEGU.png')
```
