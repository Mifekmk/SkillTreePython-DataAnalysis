```python
# 분야별 예산구조 재정추이
df = pd.read_csv("공공질서 및 안전분야 재정구조.csv", encoding='cp949')
df.shape
```

```python
df_path =r'C:\Users\HKIT\Desktop\분야별 재정지출'
```

```python
fields = os.listdir(df_path)
fields =['공공질서 및 안전 분야',
 '과학기술 분야',
 '교육 분야',
 '교통 및 물류 분야',
 '국방 분야',
 '국토 및 지역개발 분야',
 '농림수산 분야',
 '문화 및 관광 분야',
 '보건 분야',
 '사회복지 분야',
 '산업·중소기업 및 에너지 분야',

 '일반·지방행정 분야',
 '통신 분야',
 '통일·외교 분야',
 '환경 분야']
```

```python
dfs = []
for field in fields:
    print(field, end=' :')

    f = f'{df_path}/{field}/{field} 재정구조.CSV'
    print(f)
    print(os.path.isfile(f))
    if os.path.isfile(f):
        dfs.append(pd.read_csv(f, encoding='CP949'))
        print('Y')
    else: print('N')    
```

```python
for field, df in zip(fields, dfs):
    print(field)
    print(df)
```

```python
df = pd.concat(dfs, axis=0, ignore_index=True)
df.head()
```

```python
fields
```

```python
fields = ['국토 및 지역개발',
          '국방',
          '보건',
          '교육',
          '과학기술',
          '문화 및 관광',
          '사회복지', 
          '농림수산',
          '교통 및 물류',
          '공공질서 및 안전',
          '통신',
          '일반 지방행정',
          '환경',
          '산업 중소기업 및 에너지',
          '통일 외교'
          ]
df['분야'] = [field for field in fields for i in range(7)]
df.head()
```

```python
df.info()
```

```python
gru = df.groupby('연도').sum().iloc[1:]
print(gru)
gru.plot(marker='o')
```

```python
piv = pd.pivot_table(data=df, index='분야', columns='연도', values='합계예산(억원)', sort=False)
piv.style.background_gradient()
```

```python
_=sns.relplot(data=df, x='연도', y='합계예산(억원)', 
            hue='분야',marker='o', kind='line', #palette='tab20',
            aspect=2, )
```

```python
fig, axes = plt.subplots(2,3,figsize=(25,10))
axes = axes.flatten()
for y,ax in zip(range(2015,2021), axes):
  ax.set_title(f'{y} 분야별 합계예산')
  ax.pie(piv.loc[:,y], labels=fields, autopct='%.1f%%')
```
