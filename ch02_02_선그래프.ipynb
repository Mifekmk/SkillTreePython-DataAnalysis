### **02. 눈금 공유**
---



````python
t = np.arange(0.01, 5 ,0.01) # 0.01부터 5까지 0.01 단위마큼 증가
t
s1 = np.sin(t*np.pi)
s2 = np.sin(t**np.pi)
s3 = np.exp(-t)

plt.subplot(3,1,1)
plt.plot(t,s1)
plt.subplot(3,1,2)
plt.plot(t,s2)
plt.subplot(3,1,3)
plt.plot(t,s3)
````

````python
t = np.arange(0.01, 5 ,0.01) # 0.01부터 5까지 0.01 단위마큼 증가
t
s1 = np.sin(t*np.pi)
s2 = np.sin(t**np.pi)
s3 = np.exp(-t)

ax1=plt.subplot(3,1,1)
plt.plot(t,s1)
plt.tick_params('x', labelbottom=False) # 눈금지우기

plt.subplot(3,1,2, sharex=ax1)  # sharex로 x축을 공유
plt.plot(t,s2)
plt.tick_params('x', labelbottom=False) # 눈금지우기

plt.subplot(3,1,3, sharex=ax1) # 축 공유
plt.plot(t,s3)
# 마지막에는 눈금을 지우지 않습니다.
````

### **02. 램덤값 비교하기**
---

- 넘파이로 표준정규분포를 따르는 50개의 데이터를 총 세개 추출하세요.
- 세개의 데이터를 3 x 1 플롯으로 그리세요
- 세개의 플롯의 눈금을 모두 공유하세요.

````python
a = np.random.random(50)

s1 = a
s2 = a*np.sin(a)
s3 = a*np.cos(a)

ax1=plt.subplot(3,1,1)
plt.plot(a,s1)
plt.tick_params('x', labelbottom=False) # 눈금지우기

plt.subplot(3,1,2, sharex=ax1)  # sharex로 x축을 공유
plt.plot(a,s2)
plt.tick_params('x', labelbottom=False) # 눈금지우기

plt.subplot(3,1,3, sharex=ax1) # 축 공유
plt.plot(a,s3)
# 마지막에는 눈금을 지우지 않습니다.
````
