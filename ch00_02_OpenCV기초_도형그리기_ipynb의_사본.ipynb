## 이미지 준비
---

````python
path = 'sea.png' # 경로 설정
img = cv2.imread(path) # 배열로 변환 
h , w, c = img.shape
print(f'height : {h} pixels')
print(f'width : {w} pixels')
print(f'channel : {c} pixels')
cv2_imshow(img)
````

### **| 색깔 파라미터 설정**

````python
# G, B, R : 0 ~ 255
R = 170 #@param {type:"slider", min:0, max:255, step:5}
G = 175 #@param {type:"slider", min:0, max:255, step:5}
B = 135 #@param {type:"slider", min:0, max:255, step:5}
shape = (100,100,3)
COLOR = (B, G, R) # (B, G, R) 순서로 됩니다.
bkg = np.full(shape, COLOR, np.uint(8))# 백그라운드()
cv2_imshow(bkg)
````

## 이미지 크롭
---

### **| 슬라이싱을 이용한 이미지 크롭**

````python
# X, Y, SIZE : 382, 635
croped = img[0:100, 0:100]
cv2_imshow(croped)
````

#### **크롭 파라미터 설정**

````python
# X, Y, SIZE : 635, 382
X = 320 #@param {type:"slider", min:0, max:635, step:5}
Y = 70 #@param {type:"slider", min:0, max:382, step:10}
SIZE = 65 #@param {type:"slider", min:0, max:255, step:5}
shape = (100,100,3)

croped = img[Y:Y+SIZE, X:X+SIZE] # 행:y축, 열:x축
cv2_imshow(croped)
````

## 도형 그리기
---

### **| Line 그리기**

Start와 End 점을 연결하여 직선을 그립니다.

#### `cv2.line(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지 파일
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

````python
cv2.line(img, (X,Y),(X+SIZE, Y+SIZE), COLOR, 10)
cv2_imshow(img) # 초기화가 되지 않습니다. 덮어쓰기가 가능
# 초기화를 하고 싶다면 이미지 불러오기 부터 다시 설정
````

### **| 사각형 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.rectangle(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

* Sample Code
```
img = cv2.rectangle(img, (384, 0), (510, 128), (0,255,0), 3)
```

````python
cv2.rectangle(img, (X,Y), (X+SIZE, Y+SIZE), COLOR, 5)
cv2_imshow(img)
````

### **| 원 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.circle(img, center, radian, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * center – 원의 중심 좌표(x, y)
  * radian – 반지름
  * color – BGR형태의 Color
  * thickness – 선의 두께, -1 이면 원 안쪽을 채움

* Sample Code
```
img = cv2.circle(img, (447,63), 63, (0,0,255), -1)
```

````python
cv2.circle(img, (X,Y), 50,COLOR, 5)
cv2_imshow(img)
````

### **| 텍스트 추가**

#### `cv2.putText(img, text, org, font, fontSacle, color, thickness)`
* Parameters
 * img – image
 * text – 표시할 문자열
 * org – 문자열이 표시될 위치. 문자열의 bottom-left corner점
 * font – font type. CV2.FONT_XXX
 * fontSacle – Font Size
 * color – fond color
 * thickness - 선의 두께

* Sample Code
```
cv2.putText(img, 'OpenCV', (10,500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255,255,255), 2)
```

````python
cv2.putText(img, 'OpenCV', (X,Y), cv2.FONT_HERSHEY_TRIPLEX, 2, COLOR, 4)
cv2_imshow(img)
````
