**예제 01**

다음과 같이 연도, 월, 날짜가 기입된 데이터프레임을 작성하고 `df01` 로 바인딩하세요.

2001년 1-1일 부터 2001년 2-28일까지 기입되어 있습니다. (31+28= 총 59줄)


| index | 연도 | 날짜 | 
|--|--|--|
|0|2001|1-1|
|1|2001|1-2|
|2|2001|1-3|
||...|

````python
import numpy as np
import pandas as pd
m1 = np.array([f"1-{i}" for i in range(1,32)])
m2 = np.array([f"2-{i}" for i in range(1,29)])
days = np.concatenate([m1,m2], axis=0)
years = np.array([2001 for i in range(59)])
days = days.reshape(59,1)
years = years.reshape(59,1)
table = np.concatenate([years, days], axis=1)

df01 = pd.DataFrame(table, columns=["연도", "날짜"])
df01
````

**예제 03**

`df01` 에 1월, 2월에 이어서 3월을 추가하세요.

````python
df02 = pd.DataFrame({"연도":[2022 for i in range(31)],
                    "날짜":[f"3-{i}" for i in range(1,32)]})
df01 = pd.concat([df01,df02], axis=0, ignore_index=True)
df01
````

**예제 06**

`df01`에 `요일` 열을 추가하세요. 첫번째 행의 시작은 `토요일` 입니다.

````python
weeks = ["토요일","일요일","월요일","화요일","수요일","목요일","금요일"]*(len(df01)//7)
weeks_1 = [weeks[i] for i in range(len(df01)%7)]
week = weeks+weeks_1

df_week = pd.DataFrame({"요일" : week})
df01 = pd.concat([df01,df_week], axis=1)
df01
````
