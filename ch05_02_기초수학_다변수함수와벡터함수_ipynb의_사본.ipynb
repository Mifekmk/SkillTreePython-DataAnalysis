### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x): # x = 스칼라
    return (x**2) + (2*x) + 1
f1(-1)
````

* 인풋 정의 (-10~10, 100개)

````python
# 100개 스칼라 값
xs = np.linspace(-10, 10, 100)
# 100개 스칼라 값에 대한 100게 스칼라 아웃풋
fs = f1(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, fs)
# 설명
ax.set_title('f(x) according to x', fontsize=50) # 제목
ax.set_xlabel('x',fontsize=30) # x축 라벨
ax.set_ylabel('f(x)',fontsize=30) # y축 라벨

x = 5
ax.plot(x, f1(x), marker='o', c='r') # 특정 x에 따른 f(x)를 표시
ax.text(x, f1(x), f'({x},{np.round(f1(x),2)})') # np.round 를 통해 소수점 아래 2개가 남겨지게 합니다.
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상의 모든 실수
  * 그래프의 특징 : x가 -1일 때 가장 아웃풋이 작고, x가 -1에서 멀어질 때 아웃풋이 커집니다.

#### **예제 02**

\begin{align}
f(x) = cosin(x)
\end{align}

* 함수 생성하기

````python
np.cos(np.pi/2)
````

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x,y):  # x,y 모두 스칼라
    return x+y
print(f3(1,2))
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15,25,40)
ys = np.linspace(-15,25,40)
print(xs.shape)
print(ys.shape)
X, Y=np.meshgrid(xs,ys) # 1,600개의 값이 나옵니다.
````

````python
xs = np.linspace(-15,20,36)
ys = np.linspace(-15,20,36)
print(xs.shape)
print(ys.shape)
X, Y=np.meshgrid(xs,ys) # xs, ys의 순서쌍 모두 반환

F=f3(X,Y) # X, Y 모든 쌍에 대한 함수 결과 반환 
print(F.shape)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') # 3차원 플롯
ax.scatter3D(X, Y, f3(X,Y)) # x,y,z 축 데이터를 입력 3차원 점 그리기

# 설명
ax.set_title('f(x,y) according to x,y') # 제목
ax.set_xlabel('x') # x축 라벨
ax.set_ylabel('y') # y축 라벨
ax.set_ylabel('f(x,y)')
````

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') # 3차원 플롯
ax.scatter3D(X, Y, f3(X,Y)) # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(X,Y,f3(X,Y),cmap='rainbow', alpha=0.5) # 점을 이은 surface 면 그리기
# 설명
ax.set_title('f(x,y) according to x,y') # 제목
ax.set_xlabel('x') # x축 라벨
ax.set_ylabel('y') # y축 라벨
ax.set_zlabel('x+y')

# 컬러바
fig.colorbar(surf, label='f(x,y)')
plt.show()
# x = 5
# ax.scatter3D(x, Y,f3(x,Y), marker='o', c='r') # 특정 x에 따른 f(x)를 표시
# ax.text(x, f3(x), f'({x},{np.round(f3(x),2)})') # np.round 를 통해 소수점 아래 2개가 남겨지게 합니다.
# plt.show()
plt.savefig('3D.png')
````

  * f(x)가 가질 수 있는 값 (아웃풋) : 모든 실수
  * 그래프의 특징 :

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
    return np.array([x**2,5*x])
f6(1)
````

* 인풋 정의 (-10~10 20개)

````python
xs = np.linspace(-10,10,20) # x값 20개에 따른
hs, gs = f6(xs)
# 인풋 : 첫번째 벡터(아웃풋) 요소, 두번째 벡터요소
xs.shape, hs.shape, gs.shape
````

* 2d플롯으로 시각화

````python
# 아우풋(f(x)) 스페이스 그려보기
# x축을 h(x), y축을 g(x)

fig, ax = plt.subplots(1, figsize=(6,5))
ax.plot(gs, hs) # x축을 gs, y축을 hs 선 그리기
# 설명
ax.set_title('f(x) space', fontsize=50) # 제목
ax.set_xlabel('g(x)',fontsize=30) # x축 라벨
ax.set_ylabel('h(x)',fontsize=30) # y축 라벨

scat=ax.scatter(gs,hs,alpha=0.5, c=xs, cmap='rainbow')
fig.colorbar(scat)
# x = 5
# ax.plot(x, f1(x), marker='o', c='r') # 특정 x에 따른 f(x)를 표시
# ax.text(x, f1(x), f'({x},{np.round(f1(x),2)})') # np.round 를 통해 소수점 아래 2개가 남겨지게 합니다.
plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든 실수 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 0 이상의 모든 실수

* 그래프 설명

  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) :

## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x,y):
    return np.array([x**2, y**2,x*y])
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-0.5,1,50)
ys = np.linspace(-0.5,1,50)

# x, y의 격자 순서쌍
X, Y = np.meshgrid(xs,ys)
# 아웃풋 정의
A, B, C =f9(X,Y)
# 인풋 두개, 아웃풋 세개
X.shape, Y.shape, f9(X,Y).shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') # 3차원 플롯
# 아웃풋 스페이스 (A, B, C)
ax.scatter3D(A,B,C,c='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(A,B,C,cmap='rainbow', alpha=0.7) # 점을 이은 surface 면 그리기
# 설명
ax.set_title('f(x,y) Space') # 제목
ax.set_xlabel('$x^2$') # x축 라벨
ax.set_ylabel('$y^2$') # y축 라벨
ax.set_zlabel('$xy$')
plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(x,y):
    return np.array([x,y,1+x**2+y/(1+y**2)])
f10(2,2)
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
xs = np.linspace(0,1,50)
ys = np.linspace(0,1,50)
X,Y =np.meshgrid(xs,ys)
A, B, C =f10(X,Y)
X.shape, Y.shape, f10(X,Y).shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize=(9,7))
ax = plt.axes(projection='3d') # 3차원 플롯
# 아웃풋 스페이스 (A, B, C)
ax.scatter3D(A,B,C,c='k') # x,y,z 축 데이터를 입력 3차원 점 그리기
surf = ax.plot_surface(A,B,C,cmap='rainbow', alpha=0.7) # 점을 이은 surface 면 그리기
# 설명
ax.set_title('f(x,y,z) Space') # 제목
ax.set_xlabel('$x$') # x축 라벨
ax.set_ylabel('$y$') # y축 라벨
ax.set_zlabel('$1+x**2+y/(1+y**2)$')
plt.show()
````
